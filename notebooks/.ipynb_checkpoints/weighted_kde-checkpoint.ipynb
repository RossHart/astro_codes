```python
from scipy import stats
from statsmodels.distributions.mixture_rvs import mixture_rvs
from sklearn.neighbors.kde import KernelDensity
from astroML.linear_model import NadarayaWatson
```

```python
np.random.seed(12345)

x = mixture_rvs([.25,.75], size=100, dist=[stats.norm, stats.norm],
                kwargs = (dict(loc=-1,scale=.5),dict(loc=1,scale=.5)))

w = (-0.3*x) + 1

_ = plt.scatter(x,w)
```

```python
bw = 0.3

kde = KernelDensity(kernel='gaussian',bandwidth=bw).fit(x[:,np.newaxis])
weighting_kde = NadarayaWatson(kernel='gaussian',h=bw).fit(x[:,None],w[:,None])
```

```python
x_guide = np.linspace(0,3,100)#[:,np.newaxis]

_ = plt.scatter(x,w,color='k')
w_guide = weighting_kde.predict(x_guide[:,None])#_guide)

_ = plt.plot(x, w_guide,color='g')
```

```python
def fit_clf(x,y,h):
    
    x_guide = np.linspace(np.min(x),np.max(x),1000)
    clf = NadarayaWatson('gaussian', h=h)
    clf.fit(x[:, None],y)
    y_guide = clf.predict(x_guide[:, None])
    return x_guide, y_guide
```

```python
x_, y_ = fit_clf(z_sample,mu_sample,0.1)
x_, y_ = fit_clf(x,w,0.1)
```
