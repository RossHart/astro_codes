```python
from PIL import Image
import urllib

def display_image(url_name,crop_in=100):
    # Get the url name:
    urllib.request.urlretrieve(url_name,"image.jpg")
    # Open -> crop -> display -> remove the image.
    im=Image.open("image.jpg")
    l=424 # Image size
    im=im.crop((crop_in,crop_in,l-crop_in,l-crop_in))
    plt.imshow(im)
    os.remove("image.jpg")
    plt.xticks([])
    plt.yticks([])
    return None
```

```python
data = Table.read('../../fits/full_sample_debiased_w_low_z.fits')
low_z_cut = data['REDSHIFT_1'] <= 0.03
low_z_data = data[low_z_cut]
```

```python
def get_urls(data,question,answer,threshold=0.75,N_images=1,questions=questions):
    
    np.random.seed(1)
    if questions[question]['pre_questions'] != None:
        question_list = questions[question]['pre_questions'].copy()
    else:
        question_list = []
    if questions[question]['pre_answers'] != None:
        answer_list = questions[question]['pre_answers'].copy()
    else:
        answer_list = []
    question_list.append(question)
    answer_list.append(answer)
    
    p = np.ones(len(data))
    for i,_ in enumerate(question_list):
        p = p*data['{}_{}_debiased_rh'.format(question_list[i],answer_list[i])]
    mask = p > threshold
    
    urls = []
    jpeg_urls = data['jpeg_url'][mask]
    np.random.shuffle(jpeg_urls)
    for i in range(N_images):
        urls.append(jpeg_urls[i])
    
    return urls
```

```python
params = [['t02_edgeon','a04_yes',2],['t01_smooth_or_features','a01_smooth',4],
          ['t04_spiral','a08_spiral',5],['t03_bar','a06_bar',2],
          ['t08_odd_feature','a22_irregular',2]]

urls = []
for p in params:
    question, answer, N_images = p
    url_list = get_urls(data,question,answer,N_images=N_images)
    urls.extend(url_list)
np.random.seed(0)
np.random.shuffle(urls)
```

```python
fig,axarr = plt.subplots(3,5,figsize=(15,9))
axarr = axarr.ravel()
plt.subplots_adjust(hspace=0,wspace=0,left=0,right=1,top=1,bottom=0)
for i,u in enumerate(urls):
    plt.sca(axarr[i])
    display_image(u)

fig.text(0.99,0.01,'Image credit: Galaxy Zoo / SDSS',
         ha='right',va='bottom',color='w',size=18
         ,family='sans-serif',style='normal')

plt.savefig('blog_images.jpg')
```
